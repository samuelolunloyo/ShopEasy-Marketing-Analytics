```python
# import all necessary libraries
import pandas as pd
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer as SIA
```

```python
# lets load our table into python
df = pd.read_csv('Documents\CustomerReviews.csv')
# Remove Null rows in the date column and set datatypes
df = df.dropna(subset = ['ReviewDate'])
df['ReviewText'] = df['ReviewText'].astype(str)     # Converts ReviewText to String format
```

```python
# Calculate Sentiment score
analyzer = SIA()
df['compound'] = [analyzer.polarity_scores(x) ['compound'] for x in df['ReviewText']]
df['negative'] = [analyzer.polarity_scores(x) ['neg'] for x in df['ReviewText']]
df['neutral'] = [analyzer.polarity_scores(x) ['neu'] for x in df['ReviewText']]
df['positive'] = [analyzer.polarity_scores(x) ['pos'] for x in df['ReviewText']]
```

```python
def categorize_sentiment(row):
    score = row['compound']
    rating = row['Rating']
    
    # Adjust score based on rating
    if rating >= 4:
        score += 0.1
    elif rating <= 2:
        score -= 0.1
    
    # Categorize sentiment
    if score >= 0.5:
        return 'Positive'
    elif score >= 0.05:
        return 'Somewhat Positive'
    elif score > -0.05:
        return 'Neutral'
    elif score > -0.5:
        return 'Somewhat Negative'
    else:
        return 'Negative'
    
# Apply to DataFrame
df['sentiment_analysis'] = df.apply(categorize_sentiment, axis=1)

# Export the updated DataFrame to the original file (or a new one)
df.to_csv('Documents\CustomerReviews.csv', index=False)
```
